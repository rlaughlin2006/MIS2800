```python
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
%%sql
SELECT * FROM Marketing_Brands;
```

```python
%%sql
SELECT brand_name, brand_type FROM Marketing_Brands;
```

```python
%%sql
SELECT brand_name, brand_type FROM Marketing_Brands WHERE brand_type = 'Religious';
```

```python
%%sql
SELECT brand_name FROM Marketing_Brands WHERE brand_budget > 700000;
```

```python
%%sql
SELECT * FROM Marketing_Brands WHERE brand_budget > 7000000
```

```python
%%sql
SELECT brand_name FROM Marketing_Brands WHERE brand_type IN ('Snacks', 'Beverages');
```

```python
%%sql
SELECT * FROM Marketing_Brands ORDER BY brand_budget DESC;
```
